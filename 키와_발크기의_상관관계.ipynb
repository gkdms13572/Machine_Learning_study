```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 데이터 수집

```python
 # 데이터 저장
 height = [155.3, 157.5, 156.5, 163.9, 169.3,170.5, 195.5, 175.1,173.8,177.7, 182.6,180.7, 186.7, 189.9, 189.2]
 foot_size = [220,230,235,240,245, 250,255,260,265,270,275,280,285,290,300]
 print(len(foot_size), len(height))
```

```python
# 데이터 그래프로 시각화하기
plt.scatter(foot_size, height)
plt.xlabel('Foot Size')
plt.ylabel('Height')
plt.show()
```

- 190cm키에 다른 데이터에 비해 완전 벗어난 것이 보임 -> 극단치
- 극단치를 처리해야 인공지능 모델의 정확도를 높일 수 있다

# 데이터 전처리

```python
# 극단치 데이터 제거
# 극단치 데이터 제거하기 위해 0~5번 + 7~ 14번 (극단치 레이블이 6이라)
foot_size = foot_size[0:6] + foot_size[7:15]
height = height[0:6] + height[7:15]
```

```python
plt.scatter(foot_size, height)
plt.xlabel('Foot Size')
plt.ylabel('Height')
plt.show()
```

# 회귀 분석

- 회귀 분석은 머신러닝의 기본 바탕이고, 오차가 가장 작은 선을 찾는 것이 중요함.
- 오차가 가장 작은 선을 찾는 방법 : 평균제곱오차의 미분 값이 0인 a(기울기)와 b(절편)를 구하는 것임. -> 최소제곱법

```python
# 최소제곱법을 구현할 수 있게 넘파일 배열로 데이터를 변환해서 불러옴
# 기본 배열(array)은 연산할 때 불편하고 안 되는 것들이 많음
```

```python
# 넘파일 배열 변환
x = np.array(foot_size)
y = np.array(height)
```

```python
# 최소제곱법 구하는 방법
# y - y의 평균 곱하기 x-x의 평균 곱해서 1부터 n까지 더하기 -> 분자
a = np.sum((y-np.mean(y))*(x-np.mean(x)))
# x - x의 평균 제곱을 1부터 n까지 더하기 -> 분모
a = a/np.sum((x-np.mean(x))**2)
b = np.mean(y) - a*np.mean(x)
print('a(기울기):',a, 'b(절편)',b)
```

```python
# 회귀선 구하기
line_x = np.arange(min(x), max(x),1) # x의 최솟값과 최댓값 사이에 수를 1cm 간격으로 line_x에 저장해줌
line_y = a*line_x + b # line_x의 데이터를 우리가 구한 직선에 대입해서 그 값을 line_y에 저장
```

```python
# 그래프 그리기
plt.plot(line_x, line_y, color='red')
plt.scatter(x,y)
plt.xlabel('Foot Size')
plt.ylabel('Height')
plt.show()
```
